```python
!pip install requests beautifulsoup4
import requests
from bs4 import BeautifulSoup
url = input("enter url")
derinlik = int(input("tarama derinliği gir(  0-3)"))
def web_sayfasi_tara(url, derinlik):
    if derinlik == 0:
        return
    response = requests.get(url)
    html_icerigi = response.content
    soup = BeautifulSoup(html_icerigi, "html.parser")
    title = soup.title.string if soup.title else "Başlık bulunamadı"
    with open("tarama_sonuclari.txt", "a") as f:
        f.write(f"URL: {url}\nBaşlık: {title}\n\n")
    linkler = soup.find_all("a")
    for link in linkler:
        print("bağlantı:", link.get("href"))

web_sayfasi_tara(url, derinlik)


```
